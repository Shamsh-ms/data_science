```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#generating data
data=np.linspace(-10,10,200)
```

```python
def sigmoid(data):
    val=1/(1+np.exp(-data))
    return val
```

```python
output=sigmoid(data)
```

```python
plt.plot(data,output)
plt.xlabel("data")
plt.ylabel("output")
plt.title("Sigmoid Function")
```
